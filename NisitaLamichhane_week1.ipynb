```python
#Problem - 3: Array Operations
import numpy as np

x = np.array([[1,2],[3,5]])
y = np.array([[5,6],[7,8]])
v = np.array([9,10])
w = np.array([11,12])

#1
print(np.add(x, y))
print(np.add(v, w))
print(" ")

#2
print(np.subtract(x, y))
print(np.subtract(v, w))
print(" ")

#3
print(np.multiply(x,2))
print(" ")

#4
print(np.square(y))
print(" ")

#5
print("v.w =", np.dot(v, w))
print("x.v =", np.dot(x, v))
print("x.y =", np.dot(x, y))
print(" ")

#6
xy_concatenated = np.concatenate((x, y), axis=0)
vw_concatenated = np.concatenate((v.reshape(-1, 1), w.reshape(-1, 1)), axis=1)
print("Concatenated x and y along rows:")
print(xy_concatenated)
print("Concatenated v and w along columns:")
print(vw_concatenated)

#7
print(np.concatenate(x, v))
```

The error occurs because np.concatenate expects a list of arrays as its first argument, followed by an optional axis argument. Here, the  provided x (an array) as the first argument and v (another array) as the second. NumPy is mistakenly interpreting v as the axis argument, which should be an integer, not an array.

```python
#Problem - 4: Matrix Operations
import numpy as np
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Verify A * A^-1 = I
inv_A = np.linalg.inv(A)
identity_matrix = np.eye(2)
lhs_1 = np.dot(A, inv_A)

print("1. Verifying A * A^-1 = I")
print("A * A^-1 = \n", lhs_1)
print("Identity Matrix = \n", identity_matrix)

if np.allclose(lhs_1, identity_matrix):
    print("Result: A * A^-1 = I (Proven)\n")
else:
    print("Result: A * A^-1 ≠ I\n")

# 2. Verify AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)

print("2. Verifying AB ≠ BA")
print("AB = \n", AB)
print("BA = \n", BA)

if np.array_equal(AB, BA):
    print("Result: AB = BA (Unexpected)\n")
else:
    print("Result: AB ≠ BA (Proven)\n")

# 3. Verify (AB)^T = B^T * A^T
AB_T = np.transpose(np.dot(A, B))
BT_AT = np.dot(np.transpose(B), np.transpose(A))

print("3. Verifying (AB)^T = B^T * A^T")
print("(AB)^T = \n", AB_T)
print("B^T * A^T = \n", BT_AT)

if np.allclose(AB_T, BT_AT):
    print("Result: (AB)^T = B^T * A^T (Proven)\n")
else:
    print("Result: (AB)^T ≠ B^T * A^T\n")
```
