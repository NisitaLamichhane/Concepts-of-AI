```python
#To - Do - 3:
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/Concepts of AI /Datasets/student.csv")

#1
train_ratio = 0.8
train_size = int(len(data) * train_ratio)

train_data = data.iloc[:train_size]
test_data = data.iloc[train_size:]

#2
X_train = train_data.drop('Math', axis=1)
y_train = train_data['Math']
X_test = test_data.drop('Math', axis=1)
y_test = test_data['Math']
print(f"Size of Training Data: {len(X_train)}")
print(f"Size of Test Data: {len(X_test)}")
```

```python
#To - Do:
import numpy as np
import pandas as pd

def gradient_descent(X, y, W, learning_rate, num_iterations):
    m = len(y)
    for _ in range(num_iterations):
        predictions = X.dot(W)
        errors = predictions - y
        gradient = (1 / m) * X.T.dot(errors)
        W -= learning_rate * gradient
    return W

def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))

def r2_score_custom(y_true, y_pred):
    total_variance = np.sum((y_true - np.mean(y_true)) ** 2)
    unexplained_variance = np.sum((y_true - y_pred) ** 2)
    return 1 - (unexplained_variance / total_variance)

def train_and_evaluate_model(data, target_column, learning_rate=0.01, num_iterations=1000, train_ratio=0.8):
    X = data.drop(columns=[target_column]).values
    y = data[target_column].values.reshape(-1, 1)
    X = (X - X.mean(axis=0)) / X.std(axis=0)
    X = np.c_[np.ones((X.shape[0], 1)), X]
    train_size = int(len(data) * train_ratio)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]
    W = np.zeros((X_train.shape[1], 1))
    W = gradient_descent(X_train, y_train, W, learning_rate, num_iterations)
    y_pred = X_test.dot(W)
    rmse_value = rmse(y_test, y_pred)
    r2_value = r2_score_custom(y_test, y_pred)
    print(f"RMSE: {rmse_value:.4f}")
    print(f"R²: {r2_value:.4f}")
    return W

data = pd.read_csv("/content/drive/MyDrive/Concepts of AI /Datasets/student.csv")
W = train_and_evaluate_model(data, target_column='Math', learning_rate=0.01, num_iterations=1000)
```

An R² value of 0.67 indicates that the model has moderate predictive ability. While it is not flawless, it offers a reasonable approximation of the data.

The RMSE of 8.06 reflects a moderate level of error between the predicted and actual values. However, without understanding the range or scale of the target variable, it’s challenging to determine whether this level of error is significant or acceptable.

```python
#2. experiment with different value of learning rate
learning_rates = [0.001, 0.01, 0.1, 0.5, 1]
for lr in learning_rates:
    print(f"Evaluating model with learning rate: {lr}")
    W = train_and_evaluate_model(data, target_column='Math', learning_rate=lr, num_iterations=1000)
    print("-" * 50)
```

Low Learning Rate (e.g., 0.001): Training will be slower, but the model is less likely to overshoot and can converge more steadily. However, if the model performs poorly, it may be because it hasn't learned enough due to the slow updates.

Moderate Learning Rate (e.g., 0.01): Offers a balance between speed and convergence, making it a good starting point for many models.

High Learning Rate (e.g., 0.1, 0.5, 1): Training may be faster, but there's a risk of overshooting the optimal weights. If the learning rate is too high, the model might not converge or could result in large errors (high RMSE).
